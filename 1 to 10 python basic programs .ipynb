### 2] Write a Python program to do arithmetic operations addition and division.

```python
# Arithmetic Operations for addition and division
num1 = int(input("Enter a first number :"))
num2 = int(input("Enter a second number :"))

Sum = num1 + num2
print(f"Sum: {num1} + {num2} = {Sum}")


if num2 == 0:
    print("Error: Division of Zero is not allowed")
else:
    Div = num1 / num2
    print(f"Div: {num1} / {num2} = {Div}")
```

### 3] write a Python program to find the area of a triangle

```python
base = float(input("Enter a base of a triangle: "))
height = float(input("Enter a height of a triangle: "))
Area = 0.5 * base * height
print(f"Area of Triangle is {Area}")
```

### 4] Write a Python program to swap two variables.




```python
a = float(input("Enter a first number: "))
b = float(input("Enter a second number: "))
print(f"Original number a={a}, b={b}")
temp = a 
a = b
b = temp
print(f"Swapped number a={a}, b= {b}")
```

### 9] Write a Python program to Counting VOWELS in a given word(Take the input from the user).

```python
vowels = ['a', 'e', 'i', 'o', 'u']
word = input("Enter a Word:")
count = 0
for character in word:
    if character in vowels:
        count+=1
print(count)
```

### 10] Write a Python program to Counting CONSONANTS in a given word(Take the input from the user).

```python
vowels = ['a', 'e', 'i', 'o', 'u']
word = input("Please a enter a word:")
count= 0
for character in word:
    if character not in vowels:
        count+=1
print(count)
```
